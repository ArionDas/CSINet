### Input Text

```python
df['text_tokens']
```

```python
def extract_input_text(df, input_text):
    
    for num_rows in range(len(df)):
        list_ = df['text_tokens'][num_rows]
        ##print(list_)
        inp_txt = ""
        for i in range(len(list_)):
            word = list_[i].replace('\'', '').replace('"', '').replace(',', '')
            inp_txt += word
        
        inp_txt = inp_txt[1:-1]
        input_text.append(inp_txt)
            
    return input_text
        
```

```python
input_text = [] ## list for storing the input texts

input_text = extract_input_text(df, input_text)
```

```python
df['input'] = input_text
```

### Output Text

```python
def extract_claims(df, claims):
    
    for num_rows in range(len(df)):
        list_ = df['text_tokens'][num_rows]
        list_ = list_.split(", ")
        list_[0] = list_[0][1:]
        list_[-1] = list_[-1][:-1]
        claim = df['claim_vector'][num_rows]
        claim = claim.split(", ")
        claim[0] = claim[0][1:]
        claim[-1] = claim[-1][:-1]
        ##print(list_)
        out_txt = ""
        for i in range(len(list_)):
            if claim[i] == "1":
                word = list_[i][1:-1]
                out_txt += (word + ' ')
        if(out_txt==""):
            out_txt = ""
        else:
            out_txt = out_txt[:-1]
        claims.append(out_txt)
            
    return claims
```

```python
claims = []

claims = extract_claims(df, claims)
```

```python
df['output'] = claims
```
